```python
"""read in the data"""
boston_1 = pd.read_csv('boston_weather_1.csv')
boston_2 = pd.read_csv('boston_weather_2.csv')
boston_3 = pd.read_csv('boston_weather_3.csv')
boston_4 = pd.read_csv('boston_weather_4.csv')
more_variables = pd.read_csv('more_weather_variables.csv')
```

```python
"""concatenate the dataframes"""
combined_boston = pd.concat([boston_1, boston_2, boston_3, boston_4], ignore_index=True)
```

```python
"""merge (inner join) combined_boston and more_variables on indices"""
boston_data = pd.merge(combined_boston, more_variables, how='inner',
                       left_index=True, right_index=True)
```

Imagine that you wanted to use the student_data in question 2a to make predictions such that the ice_cream_flavor, math and reading columns are input variables and science column is the output variable you want to predict

a) Use the LabelBinarizer() in the sklearn package to transform the ice_cream_flavor column in the student_data to dummy variables, then join these dummy variables to the student_data and drop the original ice_cream_flavor column. Reassign the resulting DataFrame to a variable called student_data_1. Print out the entire student_data_1 DataFrame 

```python
"""
apply sklearn LabelBinarizer to student_data ice_cream_flavor column
apppend one hot encoded variable to student_data and assin to a new variable student_data_one
"""
lb = LabelBinarizer()
encoded_flavor = lb.fit_transform(flavor)
encoded_flavor = pd.DataFrame(data=encoded_flavor,
                              columns=['chocolate', 'strawberry', 'vanilla'])
student_data_1 = pd.merge(student_data, encoded_flavor, how='inner', left_index=True,
                          right_index=True)
student_data_1.drop(columns=['ice_cream_flavor'], inplace=True)
student_data_1.head(student_data_1.shape[0])
```

b) Extract the math, reading and science scores and use the StandardScaler() class in the sklearn.preprocessing module to standardize these scores, then merge the standardized scores to the dummy variables, and call the resulting DataFrame student_data_std. Print out the entire student_data_1 DataFrame.

```python
"""
apply sklearn StandardScaler to student scores (math, reading, science)
append to one hot encoded ice cream flavors and assign as dataframe to var student_data_std
"""
sc = StandardScaler()
student_data_std = student_data_1.copy()
student_data_std[['math', 'reading', 'science']] = sc.fit_transform(student_data_1[['math',
                                                                     'reading', 'science']])
student_data_std.head(student_data_std.shape[0])
```

d) Using the boston_1 DataFrame in question 1a, select the LowTemp, HighTemp, WarmestMin, ColdestHigh, AveMin, AveMax columns and use the “pipeline” functionality in sklearn to transform the selected data using the MinMaxScalar() and Imputer() classes.  With the Imputer() class, the missing data in each column should be imputed using the mean value for the column. Assign the resulting DataFrame to a variable called pipeline_data. Print the pipeline_data

```python
"""instantiate pipeline components and feed to sklearn pipeline"""
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
scaler = MinMaxScaler()
pipeline = Pipeline([('impute', imputer), ('scaler', scaler)])
pipeline
```

```python
"""apply pipeline to select columns from boston_1 dataframe"""
pipeline_data = pipeline.fit_transform(boston_1[['LowTemp', 'HighTemp', 'WarmestMin', 'ColdestHigh',
                                                 'AveMin', 'AveMax']])
pipeline_data = pd.DataFrame(pipeline_data, columns=['LowTemp', 'HighTemp', 'WarmestMin', 'ColdestHigh',
                                                     'AveMin', 'AveMax'])
pipeline_data
```
