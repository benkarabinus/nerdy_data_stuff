```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
sns.set_style('darkgrid')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.pairplot(tips)
```

```python
sns.boxplot(x="day", y="total_bill",
            hue="smoker", palette=["m", "g"],
            data=tips)
sns.despine(offset=10, trim=True)
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.histplot(
    tips,
    x="size", y='total_bill'

)
```

```python
corr = tips.corr()
sns.heatmap(corr)
```

```python
fig = plt.figure(figsize=(12, 9))
gs = fig.add_gridspec(3, 4)
ax0 = plt.subplot(gs[:2, :2])
ax1 = plt.subplot(gs[2:, :2])
ax2 = plt.subplot(gs[:4, 2:])
sns.boxplot(x="day", y="tip",
            hue="smoker", palette=["m", "g"],
            data=tips, ax=ax0)


sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', ax=ax1)

sns.heatmap(corr, ax=ax2)

fig.suptitle('Tips Data (Exploratory Analysis)', fontsize=18)
plt.tight_layout()
```
