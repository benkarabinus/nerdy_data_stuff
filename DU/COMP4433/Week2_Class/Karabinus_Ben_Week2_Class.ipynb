```python
import yfinance as yf
```

```python
stocks21 = yf.download('FB GOOGL AMZN AAPL', '2021-01-01', 
                     '2021-01-13', groupby = 'tickers')
```

```python
stocks22 = yf.download('FB GOOGL AMZN AAPL', '2022-01-01', 
                     '2022-01-13', groupby = 'tickers')
```

```python
tickers = ['AAPL', 'GOOGL', 'AMZN', 'FB']
for tick in tickers:
    # create new columns using multi-index to create range
    stocks21[('Range', tick)] = stocks21[('High',tick)] - stocks21[('Low',tick)]
    stocks22[('Range', tick)] = stocks22[('High',tick)] - stocks22[('Low',tick)]
```

```python
stocks21 = stocks21.stack(level=0).rename_axis(['Date', 'Ticker']).reset_index(level=1)
stocks21['day_of_year']
```
